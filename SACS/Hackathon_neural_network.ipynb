```python
#data preprocessing
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from sklearn.utils import shuffle

dataset = pd.read_csv("C:/Users/Nill/Desktop/datasets/data.csv")
dataset.head()
```

```python
#we create independent and dependent variable vector or feature matrix

#independent vairable
X = dataset.iloc[:, 2:32].values
#dependent variable
Y = dataset.iloc[:, 1].values
```

```python
print(X)
```

```python
print(Y)
```

```python
#encoding categorical data
from sklearn.preprocessing import LabelEncoder
labelencoder_Y = LabelEncoder()

#apply labeEncoder object on our column (0 is the index of categorical column)
Y = labelencoder_Y.fit_transform(Y)
print(Y)
```

```python
# Splitting the dataset into the Training set and Test set
from sklearn.model_selection import train_test_split

#0.2 (114 observation out of 570) will be in test set
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size = 0.2)
```

```python
from sklearn.preprocessing import StandardScaler

#create StandaradScaler object
sc_X = StandardScaler()

#fit it to the train dataset
X_train = sc_X.fit_transform(X_train)
#no need to fit to the test set (since its already fitted in train set)
X_test = sc_X.transform(X_test)
```

```python
print(X_train)
```

```python
print(X_train.shape)
print(Y_train.shape)
print(X_test.shape)
```

```python
#building the CNN

import keras
from keras.models import Sequential
from keras.layers import Dense
```

```python
#create a classifier object
classifier = Sequential()

#first hidden layer
classifier.add(Dense(output_dim = 16,  init ='uniform', activation = 'relu', input_dim = 30 ))

classifier.add(Dense(output_dim = 16,  init ='uniform', activation = 'relu'))

#final output layer
classifier.add(Dense(output_dim = 1,  init ='uniform', activation = 'sigmoid'))
```

```python
#compiling the ANN
classifier.compile(optimizer = 'adam', loss = 'binary_crossentropy', metrics = ['accuracy'])
```

```python
#fitting the ANN to the training set
classifier.fit(X_train, Y_train, batch_size = 10, nb_epoch =  100 )
```

```python
#making predictions

Y_pred = classifier.predict(X_test)
print(Y_pred)
```
